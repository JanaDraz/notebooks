```python
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt
%matplotlib inline

#003LV Lotka Volterra

b=0.2 #beta
c=b*b

bmin=0.1 #0.17
bmax=0.2 #0.34

minx = 0.
miny = 0.
maxx = 0.5
maxy = 1.

# The initial position is (0, 0).
x0 = np.zeros(4)
# The initial x and derx
x0[0] = 0.3
x0[1] = 0.7
x0[2] = x0[0]*(0.1 - b*x0[1])
x0[3] = -x0[1]*(c - 0.2*x0[0])

#f returns [x',y',x'',y''] dep on x,y,x',y',t,b
def f(x, t0, b):
    # x has four components: x=[u, u'].
    u, udot = x[:2], x[2:] #slice the array
    # We compute the second derivative u'' of u.
    udotdot = np.zeros(2)
    c = b*b
    udotdot[0] = udot[0]*(0.1-b*u[1]) + u[0]*(-b*udot[1])
    udotdot[1] = -udot[1]*(c-0.2*u[0])+(-u[1])*(-0.2*udot[0])
    # We return v'=[u', u''].
    return np.r_[udot, udotdot]

fig, ax = plt.subplots(1, 1, figsize=(8, 4))

# We want to evaluate the system on 30 linearly
# spaced times between t=0 and t=3.
t = np.linspace(0., 10., 50)

# We simulate the system for different values of k.
for b in np.linspace(bmin, bmax, 5):
    # We simulate the system and evaluate $v$ on the
    # given times.
    v = spi.odeint(f, x0, t, args=(b,))
    # We plot the particle's trajectory.
    ax.plot(v[:, 0], v[:, 1], 'o-', mew=1, ms=8,
            mec='w', label=f'b={b:.2f}') #o- x- |- .-
ax.legend()
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)

arrowcount = [8,5]

b=bmin

X = np.arange(minx, maxx, (maxx-minx)/arrowcount[0])
Y = np.arange(miny, maxy, (maxy-miny)/arrowcount[1])
U, V = np.meshgrid(X, Y)
A = U*(0.1-b*V)
B = -V*(0.05-0.2*U)

#fig, ax = plt.subplots()
q = ax.quiver(X, Y, A, B)

b=bmax
X = np.arange(minx, maxx, (maxx-minx)/arrowcount[0])
Y = np.arange(miny, maxy, (maxy-miny)/arrowcount[1])
U, V = np.meshgrid(X, Y)
A = U*(0.1-b*V)
B = -V*(c-0.2*U)

#fig, ax = plt.subplots()
q = ax.quiver(X, Y, A, B)

plt.show()
```
